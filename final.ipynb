```python
# --- Imports from your codebase ---
import yaml

from src.data.data_loader import load_train, load_test
from src.features.preprocessing import apply_preprocessing
from src.models.train_model import train_model
from src.models.predict import predict_model
from src.utils.save_submission import save_submission
from src.utils.logging_utils import get_logger
from src.utils.timer import Timer

# seeds are set inside the config file
```

```python
# note: when running the main.py you can chose which model
CONFIG_PATH = "config/model_rt_best.yaml"

with open(CONFIG_PATH, "r") as f:
    cfg = yaml.safe_load(f)

cfg

# all of the models available
# config/model_rf.yaml\n"
# config/model_svm.yaml\n"
# config/model_xgb.yaml\n"
# config/model_lr.yaml\n"
# config/model_lsvm.yaml\n"
# config/model_rt.yaml\n"
# config/model_stack.yaml\n",
```

```python
logger, timestamp = get_logger()
logger.info(f"Loaded config {CONFIG_PATH}" + yaml.dump(cfg, sort_keys=False))

# ---------------------------
# Load training & test data
# ---------------------------
with Timer("Loading training data..."):
    X_train, metadata_train, y_train = load_train()

with Timer("Loading test data..."):
    X_test, metadata_test = load_test()

# ---------------------------
# Preprocessing
# ---------------------------
with Timer("Preprocessing..."):
    X_train_prep, X_test_prep = apply_preprocessing(
        X_train=X_train,
        X_test=X_test,
        y_train=y_train,
        metadata_train=metadata_train,
        metadata_test=metadata_test,
        cfg=cfg
    )

# ---------------------------
# Train model
# ---------------------------
with Timer("Training model..."):
    model = train_model(X_train_prep, y_train, cfg)

# ---------------------------
# Predict
# ---------------------------
with Timer("Predicting..."):
    preds = predict_model(model, X_test_prep)

# ---------------------------
# Save submission
# ---------------------------
submission_path = save_submission(
    preds, 
    metadata_test, 
    timestamp
)

submission_path
```
